```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dane = pd.read_csv('dane_dif.csv', sep = ';', decimal = ',',
                   index_col=0)
```

```python
dane['Sprzedaz calkowita'] = dane['prodA'] + dane['prodB']
```

```python
kolejnoscMiesiecy = {'styczen': 1, 'luty': 2, 'marzec': 3}
dane['Miesiac'] = pd.Categorical(dane['Miesiac'], categories=kolejnoscMiesiecy.keys(), ordered=True)
dane_posortowane = dane.sort_values(by=['Miesiac', 'dzien'])
dane_posortowane.reset_index(drop=True, inplace=True)
```

```python
# Tworzenie wykresu
plt.figure(figsize=(15, 5))
# Przypisanie odpowiedniej osi X: numeracja dni od 1 do 90 (31 + 28 + 31)
dane_posortowane['dzien_kw'] = dane_posortowane.index + 1
plt.plot(dane_posortowane['dzien_kw'], dane_posortowane['Sprzedaz calkowita'])
plt.title('Sprzeda cakowita w pierwszym kwartale')
plt.xlabel('Dzie')
plt.ylabel('Sprzeda cakowita')
plt.xticks([0, 30, 31, 58, 59, 89], ['1', '31', '1', '28', '1', '31'])
plt.yticks([0, 2,11,20,22], ["0k","2k", "11k", "20k", "22k"])
# Dodanie przerywanej linii na wysokociach 20k i 2k
plt.axhline(y=20, color='gray', linestyle='--')
plt.axhline(y=2, color='gray', linestyle='--')



plt.grid(True)
plt.show()
```
